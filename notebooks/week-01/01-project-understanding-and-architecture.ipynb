## Cell 2: Locate the Project Root and Scan the Directory Tree

The LeafGuard AI repository contains Android code, backend code, model notes, and learning materials. We can discover that structure using `os.walk()`.


```python
import os
from pathlib import Path

def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'notebooks').exists() and (candidate / '.git').exists():
            return candidate
    return start

project_root = find_project_root(Path.cwd())
print(f'Project root: {project_root}')
print()
print('=== TOP-LEVEL DIRECTORY SCAN ===')

max_depth = 2
root_depth = len(project_root.parts)
for current_root, dirs, files in os.walk(project_root):
    depth = len(Path(current_root).parts) - root_depth
    if depth > max_depth:
        dirs[:] = []
        continue
    indent = '  ' * depth
    print(f'{indent}{Path(current_root).name}/')
    for file_name in sorted(files)[:6]:
        print(f'{indent}  - {file_name}')
```

## Cell 4: Map Repository Folders to Architecture Responsibilities

A software architecture becomes more concrete when you link each folder to a role in the system.


```python
folder_map = {
    'app/': 'Android presentation tier: Activities, layouts, permissions, camera integration',
    'backend/': 'Application tier: API routes, request/response handling, cloud inference',
    'model/': 'Data + intelligence tier: model notes, label definitions, ML assumptions',
    'notebooks/': 'Learning and experimentation tier: concept exploration before coding',
    'weeks/': 'Project management + evidence collection for the course roadmap',
}

print('=== FOLDER RESPONSIBILITY MAP ===')
for folder, role in folder_map.items():
    print(f'- {folder:<12} -> {role}')
```
