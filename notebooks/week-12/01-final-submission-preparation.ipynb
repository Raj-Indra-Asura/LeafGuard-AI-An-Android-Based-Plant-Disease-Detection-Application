## Cell 1: Create a Completeness Checklist

A checklist turns a vague feeling of "almost done" into measurable progress.


````python
checklist = {
    'android_ui_complete': True,
    'camera_gallery_flow_tested': True,
    'network_prediction_flow_tested': True,
    'offline_model_documented': True,
    'database_history_verified': True,
    'evidence_collected': False,
    'presentation_ready': False,
}

completed = sum(checklist.values())
total = len(checklist)
print(f'Completion: {completed}/{total} ({completed / total:.0%})')
for item, status in checklist.items():
    print(f'- {item:<32} -> {status}')
````

## Cell 2: Generate Documentation Helper Text

Final submissions usually require concise summaries for README files, demos, or viva introductions.


````python
def generate_project_summary(project_name, focus, features):
    feature_text = ', '.join(features)
    return (
        f'{project_name} is a mobile application focused on {focus}. '
        f'Key features include {feature_text}.'
    )

summary = generate_project_summary(
    'LeafGuard AI',
    'plant disease detection for farmers and students',
    ['camera-based scanning', 'cloud or on-device inference', 'history tracking', 'disease guidance']
)
print(summary)
````

## Cell 3: Estimate APK Size Contributions

APK size is often affected by model files, image assets, and native libraries.


````python
apk_parts_mb = {
    'base_app_code': 8.5,
    'tflite_model': 4.2,
    'icons_and_images': 1.3,
    'third_party_libraries': 6.1,
}

total_size = sum(apk_parts_mb.values())
print('=== APK SIZE ESTIMATE ===')
for part, size in apk_parts_mb.items():
    print(f'{part:<22} {size:>5.1f} MB')
print(f'Total estimated APK size: {total_size:.1f} MB')
````

## Cell 4: Present a Feature Matrix as a Table

A feature matrix is useful for reports, demos, and viva preparation.


````python
feature_matrix = [
    {'Feature': 'Camera scan', 'Implemented': 'Yes', 'Tested': 'Yes', 'Evidence': 'Screenshot + log'},
    {'Feature': 'Gallery import', 'Implemented': 'Yes', 'Tested': 'Yes', 'Evidence': 'Screenshot'},
    {'Feature': 'Cloud API', 'Implemented': 'Yes', 'Tested': 'Yes', 'Evidence': 'Response sample'},
    {'Feature': 'Offline guidance', 'Implemented': 'Yes', 'Tested': 'Partial', 'Evidence': 'XML parser output'},
]

headers = list(feature_matrix[0].keys())
print(' | '.join(f'{header:<14}' for header in headers))
print('-' * 72)
for row in feature_matrix:
    print(' | '.join(f'{str(row[header]):<14}' for header in headers))
````

## Cell 5: Validate the Evidence Folder Structure

Evidence is a course requirement, so it is worth checking the folder structure programmatically.


````python
import os
from pathlib import Path

def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'weeks').exists() and (candidate / '.git').exists():
            return candidate
    return start

project_root = find_project_root(Path.cwd())
expected_week = project_root / 'weeks' / 'week-12'
evidence_dir = expected_week / 'evidence'

print(f'Checking: {evidence_dir}')
if evidence_dir.exists():
    found_files = []
    for current_root, _, files in os.walk(evidence_dir):
        for file_name in files:
            found_files.append(str(Path(current_root) / file_name))
    print('Evidence files found:', found_files if found_files else 'Folder exists but is empty')
else:
    print('Evidence folder is missing. Create it before final submission.')
````

## Parallel Kotlin Track — Week 12: Final Submission with the Dual Track

Your submission readiness checklist gains one dimension in the dual-track system: proving the two
language tracks are still exact functional twins.

### Building both APKs

```
cd android-app && ./gradlew assembleDebug          # Java track (course requirement)
cd android-app-kotlin && ./gradlew assembleDebug   # Kotlin parallel track
```

Both produce `com.leafguard` APKs from the same XML layouts, assets, and behavioral contracts.
(Install only one at a time on a device — they share the `applicationId`.)

### Consistency contract verification

Before submitting, verify against `docs/JAVA_VS_KOTLIN.md`:
- [ ] Every `.java` file has a `.kt` twin at the mirrored path (14 files)
- [ ] Room schema identical (`scan_history`, same columns) in both tracks
- [ ] Same Retrofit `@POST("predict")` multipart contract
- [ ] Same TFLite input size (224), normalization, labels, and heuristic fallback
- [ ] Same intent extras keys, share templates, notification channel/ids
- [ ] Same assets (`model.tflite`, `labels.txt`, `diseases.xml`) in both `assets/` folders

### What to present

The Java track remains the CSE 2206 course deliverable ("Java for Android development").
The Kotlin track is enrichment evidence: it demonstrates you understand the *concepts*
(Activities, Room, Retrofit, TFLite) deeply enough to express them in two languages —
concepts transfer; syntax is secondary.

**✅ Checkpoint:** in your final report, cite `docs/parallel-track/CHANGES.md` for the complete Java↔Kotlin mapping and verification results.

**⚠️ Common mistake:** describing the Kotlin track as a "migration" — it is a *parallel twin*; the Java app is unchanged and remains the primary submission.

**🔑 Key point:** a fault-free dual-track system means no orphan files, no schema drift, and no behavioral differences — verified, not assumed.
