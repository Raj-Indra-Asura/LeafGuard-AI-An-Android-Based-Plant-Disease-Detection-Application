## Cell 1: Compare Notification Importance Levels

Notification channels define how visible or interruptive a notification should be.


```python
importance_levels = [
    ('MIN', 'Silent, no status bar icon prominence'),
    ('LOW', 'Silent but visible in the shade'),
    ('DEFAULT', 'Makes sound according to user settings'),
    ('HIGH', 'Heads-up behavior for urgent content'),
]

print(f'{"Level":<10} | Description')
print('-' * 60)
for level, description in importance_levels:
    print(f'{level:<10} | {description}')
```

## Cell 2: PendingIntent Flag Bitmask Calculations

Android combines flags using bitwise OR. Python can demonstrate the same binary logic.


```python
FLAG_UPDATE_CURRENT = 1 << 0
FLAG_IMMUTABLE = 1 << 1
FLAG_CANCEL_CURRENT = 1 << 2

selected_flags = FLAG_UPDATE_CURRENT | FLAG_IMMUTABLE

print('FLAG_UPDATE_CURRENT =', FLAG_UPDATE_CURRENT)
print('FLAG_IMMUTABLE    =', FLAG_IMMUTABLE)
print('FLAG_CANCEL_CURRENT =', FLAG_CANCEL_CURRENT)
print('Combined value =', selected_flags)
print('Binary form    =', bin(selected_flags))
```

## Cell 4: Measure Distance Between Two Points

Even if the app does not do advanced geospatial work, a simple distance calculation shows how location values can become meaningful.


```python
import math

def haversine_km(lat1, lon1, lat2, lon2):
    r = 6371.0
    phi1, phi2 = math.radians(lat1), math.radians(lat2)
    d_phi = math.radians(lat2 - lat1)
    d_lambda = math.radians(lon2 - lon1)
    a = math.sin(d_phi / 2) ** 2 + math.cos(phi1) * math.cos(phi2) * math.sin(d_lambda / 2) ** 2
    return 2 * r * math.atan2(math.sqrt(a), math.sqrt(1 - a))

distance = haversine_km(23.8103, 90.4125, 23.7806, 90.4074)
print(f'Approx distance: {distance:.2f} km')
```

## Cell 5: Share Intent MIME Type Matching

Share flows depend on content type. MIME types help Android decide which apps can handle a share request.


```python
mime_examples = {
    'leaf.jpg': 'image/jpeg',
    'diagnosis.txt': 'text/plain',
    'report.json': 'application/json',
}

handler_groups = {
    'image/*': ['Gallery', 'Messaging', 'Drive'],
    'text/plain': ['Notes', 'Messaging', 'Email'],
    'application/json': ['Developer tools', 'Cloud storage'],
}

for filename, mime in mime_examples.items():
    wildcard = 'image/*' if mime.startswith('image/') else mime
    print(f'{filename:<14} -> {mime:<16} -> handlers: {handler_groups.get(wildcard, [])}')
```
