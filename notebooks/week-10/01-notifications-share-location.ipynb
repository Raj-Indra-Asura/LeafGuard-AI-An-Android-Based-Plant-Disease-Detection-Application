## Cell 1: Compare Notification Importance Levels

Notification channels define how visible or interruptive a notification should be.


````python
importance_levels = [
    ('MIN', 'Silent, no status bar icon prominence'),
    ('LOW', 'Silent but visible in the shade'),
    ('DEFAULT', 'Makes sound according to user settings'),
    ('HIGH', 'Heads-up behavior for urgent content'),
]

print(f'{"Level":<10} | Description')
print('-' * 60)
for level, description in importance_levels:
    print(f'{level:<10} | {description}')
````

## Cell 2: PendingIntent Flag Bitmask Calculations

Android combines flags using bitwise OR. Python can demonstrate the same binary logic.


````python
FLAG_UPDATE_CURRENT = 1 << 0
FLAG_IMMUTABLE = 1 << 1
FLAG_CANCEL_CURRENT = 1 << 2

selected_flags = FLAG_UPDATE_CURRENT | FLAG_IMMUTABLE

print('FLAG_UPDATE_CURRENT =', FLAG_UPDATE_CURRENT)
print('FLAG_IMMUTABLE    =', FLAG_IMMUTABLE)
print('FLAG_CANCEL_CURRENT =', FLAG_CANCEL_CURRENT)
print('Combined value =', selected_flags)
print('Binary form    =', bin(selected_flags))
````

## Cell 4: Measure Distance Between Two Points

Even if the app does not do advanced geospatial work, a simple distance calculation shows how location values can become meaningful.


````python
import math

def haversine_km(lat1, lon1, lat2, lon2):
    r = 6371.0
    phi1, phi2 = math.radians(lat1), math.radians(lat2)
    d_phi = math.radians(lat2 - lat1)
    d_lambda = math.radians(lon2 - lon1)
    a = math.sin(d_phi / 2) ** 2 + math.cos(phi1) * math.cos(phi2) * math.sin(d_lambda / 2) ** 2
    return 2 * r * math.atan2(math.sqrt(a), math.sqrt(1 - a))

distance = haversine_km(23.8103, 90.4125, 23.7806, 90.4074)
print(f'Approx distance: {distance:.2f} km')
````

## Cell 5: Share Intent MIME Type Matching

Share flows depend on content type. MIME types help Android decide which apps can handle a share request.


````python
mime_examples = {
    'leaf.jpg': 'image/jpeg',
    'diagnosis.txt': 'text/plain',
    'report.json': 'application/json',
}

handler_groups = {
    'image/*': ['Gallery', 'Messaging', 'Drive'],
    'text/plain': ['Notes', 'Messaging', 'Email'],
    'application/json': ['Developer tools', 'Cloud storage'],
}

for filename, mime in mime_examples.items():
    wildcard = 'image/*' if mime.startswith('image/') else mime
    print(f'{filename:<14} -> {mime:<16} -> handlers: {handler_groups.get(wildcard, [])}')
````

## Parallel Kotlin Track — Week 10: Notifications & Share in Kotlin

The notification-channel, bitmask-flag, and share-intent concepts above translate directly.
The Kotlin twin `utils/NotificationHelper.kt` keeps the same channel id
(`leafguard_scan_reminders`), the same notification id (1001), and the same PendingIntent flags.

### Java vs Kotlin: static helper → object

```kotlin
object NotificationHelper {
    const val CHANNEL_ID = "leafguard_scan_reminders"

    fun sendScanReminderNotification(context: Context, title: String, message: String) {
        createChannel(context)
        val intent = Intent(context, MainActivity::class.java).apply {
            flags = Intent.FLAG_ACTIVITY_NEW_TASK or Intent.FLAG_ACTIVITY_CLEAR_TASK
        }
        // ... same PendingIntent (FLAG_UPDATE_CURRENT or FLAG_IMMUTABLE), same builder
    }
}
```

Note: Java's bitwise `|` becomes Kotlin's named infix operator **`or`** — the resulting flag
integer is identical (this is exactly the bitmask arithmetic modeled earlier in this notebook).

### Share intent with apply { }

```kotlin
val shareIntent = Intent(Intent.ACTION_SEND).apply {
    type = "text/plain"
    putExtra(Intent.EXTRA_SUBJECT, getString(R.string.share_subject))
    putExtra(Intent.EXTRA_TEXT, shareText)
}
startActivity(Intent.createChooser(shareIntent, getString(R.string.share_chooser_title)))
```

The share text is built from the **same** `R.string.share_result_template` resource, so shared
messages from both apps are character-identical.

**✅ Checkpoint:** confirm the `POST_NOTIFICATIONS` runtime-permission guard (API 33+) exists in both `NotificationHelper.java` and `NotificationHelper.kt`.

**⚠️ Common mistake:** writing `|` instead of `or` in Kotlin — `|` is not a bitwise operator for Int in Kotlin.

**🔑 Key point:** notification channels, PendingIntents, and share intents are OS-level contracts; both tracks drive them with the same ids, flags, and string resources.
